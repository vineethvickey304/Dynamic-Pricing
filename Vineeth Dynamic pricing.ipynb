```python
# Time-based Dynamic Pricing
import datetime

# Example usage
base_price = 100
now = datetime.datetime.now()

# Adjust the price based on the time of day
if 8 <= now.hour < 12:
    base_price * 1.2  # Increase the price in the morning
elif 12 <= now.hour < 18:
    base_price * 1.5  # Increase the price in the afternoon
else:
    base_price  # Keep the regular price in other times

print("Time-Based Price: ", base_price)


# This code adjusts the pricing of a product based on the current time of day. The price is increased 
# during specific time ranges (morning and afternoon), and the regular price is maintained at other times. 
```

```python
# Time-based Dynamic Pricing

import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime, timedelta

# Example data
now = datetime.now()
hours = pd.date_range(now, now + timedelta(hours=24), freq='H')
base_price = 100
time_based_prices = [base_price * 1.2 if 8 <= hour.hour < 12 else base_price * 1.5 if 12 <= hour.hour < 18 else base_price for hour in hours]

# Line Chart
plt.plot(hours, time_based_prices)
plt.xlabel('Days of Month')
plt.ylabel('Prices')
plt.title('Time-Based Dynamic Pricing')
plt.xticks(rotation=45) 
plt.show()
```

```python
# Demand-based Dynamic Pricing (inverted demand)

import matplotlib.pyplot as plt
import numpy as np

# Example data
np.random.seed(0)  # For reproducibility
demand_factors = np.random.uniform(0.8, 1.5, 100)  # Random factors for demand
base_price = 100

# Inverse relationship between demand and price
prices = base_price * (1 + demand_factors)  # Increasing prices
demands = 1000 / prices  # Demand decreases as price increases

# Scatter Plot
plt.scatter(prices, demands, alpha=0.7)
plt.xlabel('Prices')
plt.ylabel('Demands')
plt.title('Inverse Relationship Between Price and Demand')
plt.show()
```

```python
# Competitive-Based Dynamic Pricing is a pricing strategy that adjusts the prices of products or services 
#based on the prices set by competitors in the market. The goal is to remain competitive while maximizing 
#revenue and profit. Here's how the Competitive-based Dynamic Pricing model typically works:

# Other factors to consider are as follows:
# Competitor Price Monitoring
# Data Analysis
# Set Pricing Rules
# Dynamic Price Calculation
# Constraint Management
# Real-time Adjustments
# Consider Other Factors
# Strategic Positioning


# Competitive-based Dynamic Pricing

import numpy as np

base_price = 90
competitor_prices = [95, 98, 105, 100, 102]  # Competitor prices for the same product
max_price_increase_percentage = 20  # Set a maximum allowed increase

avg_competitor_price = np.mean(competitor_prices)

# Set the dynamic price as a percentage increase over the average competitor price
dynamic_price_percentage = 1 + (max_price_increase_percentage / 100)
dynamic_price = avg_competitor_price * dynamic_price_percentage
    
print("Competitive-Based Dynamic Pricing: ", min(dynamic_price, base_price * dynamic_price_percentage))  # Ensure a maximum percentage increase


# This code calculates a competitive-based dynamic price for a product by considering the average 
# price of competitors and applying a percentage increase. The function ensures that the dynamic price 
# does not exceed a specified maximum percentage increase over the initial base price.
```

```python
# Competitive-based Dynamic Pricing

import matplotlib.pyplot as plt

# Competitive-based Dynamic Pricing
products = ['Product A', 'Product B', 'Product C']
base_prices = [100, 120, 90]
dynamic_prices = [110, 130, 95]

plt.figure(figsize=(10, 5))

# Bar Chart
plt.subplot(1, 2, 1)
plt.bar(products, base_prices, label='Base Prices', alpha=0.7)
plt.bar(products, dynamic_prices, label='Dynamic Prices', alpha=0.7)
plt.xlabel('Products')
plt.ylabel('Prices')
plt.title('Competitive-Based Dynamic Pricing')
plt.legend()
```

```python
# SOME ADDITIONAL DYNAMIC PRICING CONCEPTS:
    
# Location-Based Dynamic Pricing involves adjusting prices based on the geographic location of the 
# customer or the product.

%pip install geopy
import geopy.distance

# Example usage
base_price = 100
customer_location = (37.7749, -122.4194)  # San Francisco, CA
product_location = (34.0522, -118.2437)   # Los Angeles, CA

#def get_location_based_price(base_price, customer_location, product_location):
# Assume customer_location and product_location are tuples (latitude, longitude)

# Calculate the distance between customer and product locations using geopy
distance_km = geopy.distance.distance(customer_location, product_location).km

# Set a price multiplier based on distance (example: decrease price as distance increases)
price_multiplier = max(1 - (distance_km * 0.05), 0.8)  # Decrease by 5% for every km, with a minimum of 80%

# Calculate the dynamic price
dynamic_price = base_price * price_multiplier
print("Location-Based Dynamic Pricing: ", dynamic_price)


# Location-based dynamic pricing is a pricing strategy where the cost of a product or service is 
# adjusted based on the geographic location of the customer, the product, or both. This strategy 
# leverages the idea that different locations may have varying levels of demand, competition, or 
# cost of living, allowing businesses to optimize pricing for specific regions.
```

```python
# Personalized Dynamic Pricing

import random

# Example usage
base_product_price = 100
customer_profile = {"age": 28, "loyalty_level": "gold"}


# Extract relevant customer attributes
age = customer_profile.get("age", 30)
loyalty_level = customer_profile.get("loyalty_level", "silver")

# Adjust the price based on customer attributes
price_adjustment = 0

# Age-based adjustment
if age < 25:
    price_adjustment -= 0.1  # 10% discount for customers under 25

# Loyalty-based adjustment
if loyalty_level == "gold":
    price_adjustment += 0.2  # 20% premium for gold-level customers

# Generate a random personalized discount for variability
random_discount = random.uniform(-0.05, 0.05)  # Random discount between -5% and 5%

# Calculate the personalized price
personalized_price = base_price * (1 + price_adjustment + random_discount)

print("Personalized Dynamic Pricing: ", personalized_price)


# Personalized Dynamic Pricing is a strategy where the price of a product or service is customized 
# for individual customers based on their unique characteristics, behaviors, or preferences. The goal 
# is to offer a tailored pricing experience that maximizes customer satisfaction, loyalty, and 
# overall revenue for the business.
```

```python
# Dynamic Surge Pricing 

import random

# Example usage
base_product_price = 100
demand_factor = 1.2  # Adjust based on real demand factors

current_demand = random.uniform(0.8, 1.5) * demand_factor

# Define surge pricing strategy (you can customize this based on your needs)
if current_demand < 1.0:
    surge_multiplier = 1.0  # No surge pricing
elif 1.0 <= current_demand < 1.5:
    surge_multiplier = 1.5  # Moderate surge pricing
else:
    surge_multiplier = 2.0  # High surge pricing

# Calculate dynamic price based on the surge multiplier
dynamic_price = base_price * surge_multiplier

print("Dynamic Surge Pricing: ", dynamic_price)


# Dynamic Surge Pricing is a pricing strategy that adjusts prices based on the demand and supply 
# conditions in real-time. It is commonly used in industries such as ride-sharing, hospitality, and 
# events where demand can vary significantly. The goal of surge pricing is to balance supply and 
# demand, ensuring that there are enough services or products available during peak demand periods.
```

```python
# Dynamic Surge Pricing (very common for ride-sharing services)

import matplotlib.pyplot as plt

# Example data
surge_multipliers = [1, 1.5, 2, 1.2, 1.8]
base_price = 100
surge_prices = [base_price * multiplier for multiplier in surge_multipliers]

# Bar Chart for surge pricing
plt.bar(surge_multipliers, surge_prices, color='green', alpha=0.7)
plt.xlabel('Surge Multiplier')
plt.ylabel('Surge Prices')
plt.title('Dynamic Surge Pricing (Bar Chart)')
plt.show()
```

```python
# Dynamic Subscription-Based Pricing (streaming services, i.e., Netflix, Spotify, etc.)

class Subscription:
    def __init__(self, base_price, usage_factor=1.0):
        self.base_price = base_price
        self.usage_factor = usage_factor

    def calculate_subscription_cost(self):
        # Calculate the dynamic cost based on base price and usage factor
        dynamic_cost = self.base_price * self.usage_factor
        return dynamic_cost

# Example usage
base_price = 50.0
user_subscription = Subscription(base_price)

# Simulate changes in usage factor (e.g., based on user activity)
usage_factors = [1.0, 1.2, 0.8, 1.5, 1.0]

for factor in usage_factors:
    user_subscription.usage_factor = factor
    dynamic_cost = user_subscription.calculate_subscription_cost()
    print(f"Usage Factor: {factor}, Dynamic Cost: ${dynamic_cost}")
    

# Dynamic Subscription-Based Pricing refers to a pricing model where the cost of a subscription 
# service can change based on various factors, such as usage patterns, customer behavior, or market 
# conditions. Unlike fixed subscription pricing, dynamic subscription-based pricing allows for 
# flexibility in adjusting subscription costs dynamically to better align with changing circumstances.
```
